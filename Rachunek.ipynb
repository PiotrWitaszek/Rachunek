```python

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats


from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
def gauss(x, sigma=1, mu=3):
    return stats.norm.pdf(x, mu, sigma)
x = np.linspace(-50, 50, 10000) #czy jeżeli chodzi o 2 pierwsze parametry mam dowolność? i czemu jkak spróbowałem pierwszy zrobić na >3 to nie wyskoczył błąd?
plt.figure(figsize=(12, 8))

plt.plot(x, gauss(x, sigma=1, mu=3), label='odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```

```python
mu = 3
sigma = 1
s = np.random.normal(mu, sigma, 10000)
median = np.percentile(s, 50) 
q1 = np.percentile(s, 25)
q3 = np.percentile(s, 75)
print(f'Q1={q1}, median = {median}, Q3 = {q3}')
```

```python
x = np.linspace(0, 100, 100)
plt.figure(figsize=(12, 8))

plt.plot(x, gauss(x, sigma=1, mu=3), label='odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```
